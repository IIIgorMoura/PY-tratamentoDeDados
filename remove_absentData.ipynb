```python
import pandas as pd

# 1. Carregar os dados do arquivo
df = pd.read_excel("dados_ausentes.xlsx")
display(df)
```

```python
# Verificar campos vazios
display(df.isnull())
```

```python
# Contar valores ausentes
print(df.isnull().sum())
```

```python
# Opção 1 de tratamento de dados
    # Excluir linhas com dados ausentes

# criar uma copia modificada da base de dados originais
df_sem_nulos = df.dropna()

display(df, df_sem_nulos)
```

```python
# Opção 2 de tratamento de dados
    # substituir dado por "Desconhecido"

# substituindo vazios em CIDADE por Desconhecido
    # inplace=True substitui diretamente no dataframe
    # se substituir errado, já era pois o dado estará preenchido
df['Cidade'].fillna('Desconhecida', inplace=True)
display(df)
```

```python
# substituir dados ausentes SALARIO por 0
df['Salário'].fillna(0, inplace=True)
display(df)
```

```python
# Substituir dados IDADE Vazios por MEDIA DE IDADES
    # se utiliza o .mean()
df['Idade'].fillna(df['Idade'].mean(), inplace=True)
display(df)
```
