```python
import pandas as pd

df = pd.read_excel('dados_ausentes.xlsx')
display(df)
```

```python
# corrigir erros FORMATAÇÃO
    # EX: Maiuscula, Minuscula
        # utiliza CAPITALIZE em Cidades, Nomes:
df['Cidade'] = df['Cidade'].str.capitalize()
# dessa forma só a da primeira palavra é CAPITALIZED

display(df)

df['Cidade'] = df['Cidade'].str.title()
# com .TITLE() cada primeira letra de palavra será CAPITALIZED

df['Nome'] = df['Nome'].str.title()

display(df)
```

```python
# substituir Valor Categórico
    # ex: valor escrito de maneira errada

# test ex: trans cidades em sigla: São Paulo >>> sp
df['Cidade'].str.lower().replace({'são paulo': "SP", 'rio de janeiro': "RJ", 'rj': 'RJ'})
```

```python
# id tipos de dados das colunas
    # txt pode estar armazenado em STR ou OBJ
coluna_string = df.select_dtypes(include=['string', 'object']).columns.to_list()
print(coluna_string)
```

```python
# trans idade em INT
# se tiver dados vazios ele não funciona
# idade deve ser trans em INT e não FLOAT, para que a seguir apareça corretamente no bloco a seguir
df['Idade'].fillna(df['Idade'].mean(), inplace=True)

df['Idade'] = df['Idade'].astype(int)
display(df)
```

```python
# id tipos de dados COLUNAS FLOAT
coluna_float = df.select_dtypes(include=['float']).columns.to_list()
print(coluna_float)
```

```python
# clonar idade de cima e inserir no debaixo
    # ffill pega o de cima
    # bfill pega o debaixo

# df['Idade'] = df['Idade'].ffill()
# df['Idade'] = df['Idade'].bfill()
```

```python
# extrair análise e fixar
    # criando nova coluna com essa analise
    # util para aquele projeto de machine learning em que faixas são melhores do que buscar idades especificas

# exemplo: Faixa etária
df['Faixa Etária'] = pd.cut(df['Idade'], bins=[0, 18, 30, 50, 100], labels=['Jovem', 'Adulto', 'Meia-idade', 'Idoso'])
    # pd.cut separa grupos valores em faixas
    # bins são as faixas

display(df)
```

```python
# Remover duplicatas
# df = df.drop_duplicates()
# da para executar assim, ou definir em que colunas será executado
df = df.drop_duplicates(subset=['Nome'])
display(df)
```
